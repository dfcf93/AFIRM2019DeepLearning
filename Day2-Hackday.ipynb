```python
from __future__ import print_function
import sys
import os
import os.path
import csv
import re
import random
import datetime
import numpy as np
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.optim as optim
from IPython.display import display, HTML
    
NUM_NGRAPHS                     = 0
MAX_QUERY_TERMS                 = 5
DATA_DIR                        = 'data/'
DATA_FILE_NGRAPHS               = os.path.join(DATA_DIR, "ngraphs.txt")

class Featurizer:
    
    def __init__(self):
        self.__load_ngraphs()

    def __load_ngraphs(self):
        global NUM_NGRAPHS
        self.ngraphs                        = {}
        self.max_ngraph_len                 = 0
        i = 0
        with open(DATA_FILE_NGRAPHS, mode='r', encoding="utf-8") as f:
            reader                          = csv.reader(f, delimiter='\t')
            for row in reader:
                self.ngraphs[row[0]]        = i
                self.max_ngraph_len         = max(self.max_ngraph_len, len(row[0]))
                i                          += 1
        NUM_NGRAPHS                         = len(self.ngraphs)
        
    def featurize(self, s):
        feat = np.zeros((NUM_NGRAPHS, MAX_QUERY_TERMS), dtype=np.float32)
        terms = s.split(' ')[:MAX_QUERY_TERMS]
        print(terms)
        for t in range(len(terms)):
            term = '#' + terms[t] + '#'
            print(term)
            term_len = len(term)
            for k in range(term_len):
                for l in range(1, self.max_ngraph_len + 1):
                    if k + l <= term_len:
                        ngraph_idx = self.ngraphs.get(term[k : k + l])
                        if ngraph_idx != None:
                            print('ngraph: {}, ngraph id: {}'.format(term[k : k + l], ngraph_idx))
                            feat[ngraph_idx, t] += 1
        return feat
    
np.set_printoptions(threshold=np.nan)
featurizer = Featurizer()
feat = featurizer.featurize('bananananananana')
#feat = featurizer.featurize(input('query: '))
#print(feat)
```
